```python
import pandas as pd
```

```python
df = pd.read_csv("Datos/all_data.csv")

df
```

```python
df.info()
```

```python
# Cambiando a tipo numérico
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
# Filtrando el dataframe
df = df.loc[df['Quantity Ordered'] != 'Quantity Ordered']

df
```

```python
# Cambiando a tipo numérico
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
# Sacando las filas nulas
# how='all' deja caer la fila o columna si faltan todos los valores en ella. how='any' si falta cualquiera
df = df.dropna(how='all')

df
```

```python
# Agregando el campo Month

# .str[0:2] >> primeros dos caracteres  
df['Month'] = df['Order Date'].str[0:2]

df
```

```python
# Agragando el campo City
def get_city(address):
    # address.split(",") devuelve una lista (strip elimina los espacios en blanco)
    return address.split(",")[1].strip(" ")


def get_state(address):
    return address.split(",")[2].split(" ")[1]


df['City'] = df['Purchase Address'].apply(
    lambda Purchase_Address: f"{get_city(Purchase_Address)} {get_state(Purchase_Address)}"
)


df.head(3)
```

### Question 1: Cuales fueron los cinco mejores meses? 

```python
# Añadiendo el campo Sales
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

```python
df_pregunta_1 = df.groupby(['Month']).sum().sort_values(by='Sales', ascending=False)[['Sales']].head()

df_pregunta_1
```

### Pregunta 2: Cuales son las tres ciudades que más vendieron?

```python
df_pregunta_2 = df.groupby(['City']).sum().sort_values(by='Sales', ascending=False)[['Sales']].head(3)

plot_pregunta_2 = df_pregunta_2.plot(
    kind='bar',
    figsize=(12, 4),
    color='#FFD700',
)

plot_pregunta_2;
```
